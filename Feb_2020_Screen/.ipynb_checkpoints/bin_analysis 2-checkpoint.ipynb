### Create functions to connect strain and compound data to automated counts

```python
def get_strain(plate_id, well_id):
    strain = all_bins_sorted.loc[
        (all_bins_sorted['Plate_id'] == plate_id) & 
        (all_bins_sorted['Well_id'] == well_id)]['Strain_id'].values[0]
    return strain
```

```python
def get_compound(plate_id):
    compound = all_bins_sorted.loc[
        (all_bins_sorted['Plate_id'] == plate_id)]['Compound_id'].values[0]
    return compound
```

### Load and format worm location data from automated counts

```python
results_files = results_folder.glob('*.csv')

analysis_results = []
for file in results_files:
    this_experiment_df = pd.read_csv(file, index_col=' ')
    plate_id = file.stem[9:18]
    this_experiment_df['Plate_id'] = plate_id
    well_id = file.stem[19].upper()
    this_experiment_df['Well_id'] = well_id
    strain = get_strain(plate_id, well_id)
    this_experiment_df['Strain'] = strain
    compound = get_compound(plate_id)
    this_experiment_df['Compound']=compound
    analysis_results.append(this_experiment_df)

analysis_results = pd.concat(analysis_results, sort=True)
grouped_results = analysis_results.groupby(['Compound', 'Strain'])
#grouped_results.head()
```

```python
# Creating lists of all compounds and strains
compounds = analysis_results['Compound'].unique().tolist()
strains = analysis_results['Strain'].unique().tolist()
strains
```

```python
# Creating a function to add the mean CI value to each corresponding axes in the figure
def get_avg_ci(compound, strain):  
    ci = grouped.loc[
            (grouped['Compound_id'] == compound) &
            (grouped['Strain_id'] == strain)][
            'chemotaxis_index'].values[0]
    rnd_ci = round(ci, 3)
    return rnd_ci
```

### Creating histograms to map worm density to plate position

```python

font1 = {'weight': 'bold',
        'size': 12,
        }
font2 = {'weight': 'bold',
        'size': 14,
        }
sns.set(style="darkgrid")
nrows = int(len(compounds))
fig, axes = plt.subplots(nrows=nrows, ncols=4, sharex = True, figsize= (20, 50))
fig.tight_layout(pad=5.0)
for i in compounds:
    for j in strains:
        group = grouped_results.get_group((i, j))
        row = compounds.index(i)
        col = strains.index(j)
        data = group['X']
        skewness = round(skew(data),3)
        ci_val = get_avg_ci(i,j)
        axes[row, col].hist(data, bins = 15)
        axes[row, col].set_title( 'Strain = ' + j + '\nCompound = ' + i , fontdict=font2 )
        axes[row,col].set_xlim(0,3500)
        axes[row,col].set_xticks([0 , 500, 1000, 1500, 2000, 2500, 3000, 3500])
        axes[row,col].set_xlabel('Compound <------- -------> Reference')
        axes[row,col].set_ylabel('# of Worms')
        axes[row,col].set_yticks([0,50,100,150])
        axes[row, col].text(x=3500, y=100, s = 'CI = ' + str(ci_val) + '\nSkew = ' + str(skewness), ha = 'right', va = 'bottom', fontdict = font1)
        axes[row, col].axvline(3500/2, ls='--')
        
fig.savefig(base_folder.joinpath('../worm_locations_hist.pdf'))
```
