```python
v = 'Strain'
 
if v == 'Strain':
    ia_sum['UB_Strain'] =  [bkey[x] for x in ia_sum['Strain']]
elif v == 'Compound':
    ia_sum['UB_Compound'] =  [bkey[x] for x in ia_sum['Compound']]

ia_sum.head()
```

### Code for unblinding metadata sheet

```python
wells = ['A', 'B', 'C', 'D']

if v == 'Strain':
    x = 'Strain Well '
    ubx = 'UB Strain Well '
elif v == 'Compound':
    x = 'Compound Well '
    ubx = 'UB Compound Well '


for index, row in mdat.iterrows():
    for w in wells:
        blind_header = x + w
        ub_header = ubx + w
        blinded = row[blind_header]
        mdat.loc[index, ub_header] = bkey[blinded]

mdat.head()
```
