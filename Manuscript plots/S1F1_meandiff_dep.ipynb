```python
import matplotlib.pyplot as plt 
from matplotlib.collections import EllipseCollection
import seaborn as sns
import pandas as pd
import numpy as np
from math import pi
```

```python
subset = ['test', 'test_N', 'difference', 'bca_low', 'bca_high']
cx10_dat = pd.read_csv('D:/_2021_08_screen/S1fu/CX10_meandif.csv', usecols=subset)
cx10_dat =cx10_dat.sort_values('difference')
cx10_dat['Strain'] = 'CX10'
cx10_dat.head()
```

```python
subset = ['test', 'test_N', 'difference', 'bca_low', 'bca_high']
pr678_dat = pd.read_csv('D:/_2021_08_screen/S1fu/PR678_meandif.csv', usecols=subset)
pr678_dat =pr678_dat.sort_values('difference')
pr678_dat['Strain'] = 'PR678'
pr678_dat.head()
```

```python
both_strains = pr678_dat.append(cx10_dat)
both_strains['Strain'] = both_strains.Strain.astype('category')
```

```python
r = [] #Creating an empty list to hold row values
r = list(range(0,8)) # Adding row values
r = list(np.repeat(r, [12], axis=0))# Extending the list of row values for all 96 conditions

c = [] #Creating an empty list to hold column values
c = list(range(0,12)) # Adding appropriate PCR plate column values
c = c*8  # Extending the list of numbers to accomodate 96 conditions
print(len(c))
```

```python
dat = pd.read_csv('D:/_2021_08_screen/analysis/mean_diff_mm.csv', usecols=subset)
dat =dat.sort_values('difference')
comp_ordr = list(dat['test'])
comp_ordr
```

```python
b= list(cx10_dat['test'])
d = []
for comp in comp_ordr:
    if comp in b:
        d.append(comp)
d
```

```python
cx10_dat['Strain'] = cx10_dat.Strain.astype('category')
cx10_dat['test'] = cx10_dat.test.astype('category')
cx10_dat
```

```python
cx10_dat.test.cat.set_categories(d, inplace=True)
cx10_dat
```

```python
cx10_dat = cx10_dat.sort_values(['test'])
cx10_dat
```

```python
cx10_dat['Column'] = c[0:len(cx10_dat)] # Adding column values to the randomized/shuffled df of compounds
cx10_dat['Row'] = r[0:len(cx10_dat)] # Adding row values to the randomized/shuffled df of compounds
cx10_dat.head()
```

```python
pr678_dat['Strain'] = pr678_dat.Strain.astype('category')
pr678_dat['test'] = pr678_dat.test.astype('category')
pr678_dat.test.cat.set_categories(d, inplace=True)
pr678_dat = pr678_dat.sort_values(['test'])
pr678_dat
```

```python
pr678_dat['Column'] = c[0:len(pr678_dat)] # Adding column values to the randomized/shuffled df of compounds
pr678_dat['Row'] = r[0:len(pr678_dat)] # Adding row values to the randomized/shuffled df of compounds
pr678_dat.head()
```

```python
# Draw each cell as a scatter point with varying size and color
g = sns.relplot(
    data=cx10_dat,
    x="Column", y="Row", hue='difference', size=(cx10_dat['bca_high']-cx10_dat['bca_low']),
    palette="vlag", hue_norm=(-5, 5), edgecolor=".7",
    height=5, sizes=(100, 800), size_norm=(2.5, 3.75),
)

#cax = g.fig.add_axes([.92, .12, .02, .8])


# Tweak the figure to finalize
g.set(xlim=(-.5,11.5), ylim=(7.5,-.5))
g.set(xlabel="", ylabel="", aspect="equal")
g.despine(left=True, bottom=True)
g.ax.margins(.02)
g.legend.set_visible(False)
g.savefig('D:/NP_manuscript/cx10_mean_diff.png')

#g.colorbar()
```
